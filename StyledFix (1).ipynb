```python
df['Waktu Pesanan Dibuat'] = df['Waktu Pesanan Dibuat'].astype(str)

# Tangani nilai NaN dengan mengisi nilai default, misalnya dengan string kosong
df['Waktu Pesanan Dibuat'] = df['Waktu Pesanan Dibuat'].fillna('')

# Year
df['Year'] = df['Waktu Pesanan Dibuat'].apply(lambda x: x.split("-")[0] if x != '' else '')

# Month
df['Month'] = df['Waktu Pesanan Dibuat'].apply(lambda x: x.split("-")[1] if len(x.split("-")) > 1 else '')

# Day
df['Day'] = df['Waktu Pesanan Dibuat'].apply(lambda x: x.split("-")[2] if len(x.split("-")) > 2 else '')
```

```python
def convert_date(date):
    try:
        # Coba konversi dari format datetime dengan waktu
        return pd.to_datetime(date, format='%Y-%m-%d %H:%M')
    except ValueError:
        try:
            # Coba konversi dari format datetime tanpa waktu
            return pd.to_datetime(date, format='%Y-%m-%d')
        except ValueError:
            try:
                # Coba konversi dari serial Excel
                return pd.to_datetime('1899-12-30') + pd.to_timedelta(float(date), 'D')
            except ValueError:
                # Jika semua konversi gagal, return NaT
                return pd.NaT

# Terapkan fungsi konversi ke kolom 'Waktu Pesanan Dibuat'
df['Waktu Pesanan Dibuat'] = df['Waktu Pesanan Dibuat'].apply(convert_date)

# Buang baris yang gagal dikonversi (opsional)
df = df.dropna(subset=['Waktu Pesanan Dibuat'])

# Extract the month and year
df['YearMonth'] = df['Waktu Pesanan Dibuat'].dt.to_period('M')

# Aggregate transactions by month
monthly_purchases = df.groupby('YearMonth').size()

# Plotting
plt.figure(figsize=(14,7))
monthly_purchases.plot(kind='line')
plt.title('Monthly Purchases')
plt.ylabel('Number of Transactions')
plt.xlabel('Month')
plt.grid(True)
plt.show()
```

```python
# Group by TransactionID and create a list of items
transactions = df.groupby('No. Pesanan')['Nama Produk'].apply(list).reset_index(name='Nama Produk')

# Use TransactionEncoder
te = TransactionEncoder()
te_array = te.fit_transform(transactions['Nama Produk'])

# Convert to DataFrame
basket = pd.DataFrame(te_array, columns=te.columns_)

# Add TransactionID back to the DataFrame
basket['No. Pesanan'] = transactions['No. Pesanan']
```

```python
basket.head()
```

```python
# Drop the TransactionID column before applying Apriori
basket_sets = basket.drop('No. Pesanan', axis=1)

# Apply the Apriori algorithm
frequent_itemsets = apriori(basket_sets, min_support=0.0050, use_colnames=True)
```
